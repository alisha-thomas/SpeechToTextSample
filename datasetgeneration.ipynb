```python
import os

# Load and process the transcriptions file
transcriptions_file = os.path.join(os.getcwd(), 'cleaned_audios_zip', '20', '205', '20-205.trans.txt')

# Read the file and extract audio file paths and their corresponding transcriptions
audio_files = []
transcriptions = []

with open(transcriptions_file, "r") as file:
    for line in file:
        parts = line.strip().split(' ', 1)
        if len(parts) == 2:
            # audio_files.append(f"/cleaned_audios_zip/20/{parts[0]}.wav")
            path_to_audio = os.path.join(os.getcwd(), 'cleaned_audios_zip', '20', '205', f'{parts[0]}.wav')
            audio_files.append(path_to_audio)
            transcriptions.append(parts[1])

# Print the extracted audio files and transcriptions
print("Audio Files:", audio_files)
print("Transcriptions:", transcriptions)
```
